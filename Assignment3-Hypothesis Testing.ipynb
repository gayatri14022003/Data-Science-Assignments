# Hypothesis testing-Cutlets data

Q1]A F&B manager wants to determine whether there is any significant difference in the diameter of the cutlet between two units. A randomly selected sample of cutlets was collected from both units and measured? Analyze the data and draw inferences at 5% significance level. Please state the assumptions and tests that you carried out to check validity of the assumptions.
     Minitab File : Cutlets.mtw


ïƒ¨	Null hypothesis H0= mu1=mu2
Alternate hypothesis H1=mu1 != mu2
Alpha=0.05%
Significance=5%


```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import seaborn as sns
import statsmodels.api as sm
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
data=pd.read_csv("C:\\Users\\hp\\Downloads\\Cutlets.csv")
data.head()
```

```python
data.describe()
```

```python
plt.subplots(figsize=(6,3))
plt.subplot(121)
plt.boxplot(data['Unit A'])
plt.title("Unit A")
plt.plot()
plt.subplot(122)
plt.boxplot(data['Unit B'])
plt.title("Unit B")
plt.plot()
```

```python
plt.subplots(figsize=(6,3))
plt.subplot(121)
plt.hist(data['Unit A'],edgecolor='black', linewidth=2)
plt.plot()
plt.subplot(122)
plt.hist(data['Unit B'],edgecolor='black',linewidth=2)
plt.plot()
```

```python
plt.title("Unit A vs Unit B")
sns.distplot(data['Unit A'])
plt.plot()
sns.distplot(data['Unit B'])
labels=['Unit A','Unit B']
plt.legend(labels)
plt.plot()
```

```python

sm.qqplot(data['Unit A'],line= 'q')
plt.title("Unit A")
plt.plot()
sm.qqplot(data['Unit B'],line='q')
plt.title("Unit B")
plt.plot()
```

```python
statistics,p_value=stats.ttest_ind(data['Unit A'],data['Unit B'],alternative="two-sided")
print("p_ value=",p_value)
```

```python
alpha=0.025
if(alpha<=p_value):
    print("We are rejecting null hypothesis")
else:
    print("We are accepting null hypothesis")
```

```python
from scipy.stats import chi2_contingency
buyer=pd.read_csv("C:\\Users\\hp\\Downloads\\BuyerRatio.csv")
buyer.head()
```

```python
table=[[50,142,131,70],[550,351,480,350]]
```

```python
stat, p, dof, expected = stats.chi2_contingency(table)
print("stats=",stat,"\np_value=",p,"\ndoc=",dof,"\nexp=",expected)
```

```python
observed=np.array([50,142,131,70,550,351,480,350])
expected=np.array([111.01694915,91.21892655,113.05225989,77.71186441,488.98305085,401.78107345,497.94774011, 342.28813559])
```

```python
statistics,p_value=stats.chisquare(observed,expected,ddof=3)
print("statistics=",statistics,"\np_value=",p_value)
```

```python
alpha=0.05
if(alpha<=p_value):
    print("we are rejecting null hypothesis ")
else:
    print("we are accepting null hypothesis ")
```
