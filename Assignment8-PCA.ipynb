```python
#There are some quite correlation between variables. For example
#the correlation between flavanoids and dilution is pretty high (78%).
#Thus we can remove that variable from our dataset. However this method is
#long and tedious.Hence we PCA method for Dimensionality Reduction
```

```python
#dimantionality reduction with PCA
```

```python
data_norm=StandardScaler().fit_transform(data1)
```

```python
pca=PCA(n_components=13)
```

```python
principalComponents=pca.fit_transform(data_norm)
```

```python
pc=range(1,pca.n_components_+1)
plt.bar(pc,pca.explained_variance_ratio_,color='blue')
plt.xlabel('principal component')
plt.ylabel('variance%')
plt.xticks(pc)
plt.plot()
```

```python
pca_comp=pd.DataFrame(principalComponents)
```

```python
plt.scatter(pca_comp[0],pca_comp[1],alpha=0.3)
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
```

```python
#As shown in the bar graph, the most of varianve is put in the
#first 2 components. Since there is not much variance present from
#3rd component, lets just the first 2 componets in our analysis.
#The scatter plot given an indication that there may be 3 clusters
#present
```

```python
#finding optimal no of clustering
```

```python
wcss=[]
```

```python
for i in range(1,15):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(pca_comp.iloc[:,:3])
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,15),wcss)
plt.title('the elbow method')
plt.xlabel('number of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
#The scree plot levels off at k=3 and let's use it to determine 
#the clusters
```

```python
model=KMeans(n_clusters=3)
model.fit(pca_comp.iloc[:,:2])
```

```python
labels=model.predict(pca_comp.iloc[:,:2])
```

```python
plt.scatter(pca_comp[0],pca_comp[1],c=labels)
plt.show()
```
